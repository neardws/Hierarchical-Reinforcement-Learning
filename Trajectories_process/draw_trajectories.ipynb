```python
import matplotlib.pyplot as plot
import pandas as pd

csv_file = '/home/neardws/Hierarchical-Reinforcement-Learning/CSV/vehicle_1116_08_processed.csv'

df = pd.read_csv(csv_file)
x_max = df['longitude'].max()
y_max = df['latitude'].max()
plot.xlim(0, x_max)
plot.ylim(0, y_max)

chunk_size = 1800
number = 0
for chunk in pd.read_csv(csv_file, error_bad_lines=False, chunksize=chunk_size):
    colors = ['r', 'g', 'c', 'b', 'y', 'k']
    trace_id = chunk['vehicle_id'].drop_duplicates()
    for id in trace_id:
        if number >= 10:
            break
        print(id)
        trace = chunk[
            (chunk['vehicle_id'] == id)]
        if len(trace):
            x = trace['longitude']
            y = trace['latitude']
            plot.scatter(x, y, 0.1, colors[int(id) % 6])
        number += 1
plot.show()

```
